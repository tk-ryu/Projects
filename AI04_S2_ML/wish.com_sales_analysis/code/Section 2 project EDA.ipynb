```python
USD_index = full_data_drop[full_data_drop['retail_price_currency'] == 'USD'].index
print(full_data_drop.loc[USD_index, 'retail_price'].head(3))
full_data_drop.loc[USD_index, 'retail_price'] = full_data_drop.loc[USD_index, 'retail_price'] * 0.84
print(full_data_drop.loc[USD_index, 'retail_price'].head(3))
print(full_data_drop.loc[USD_index, 'discount_price'].head(3))
full_data_drop.loc[USD_index, 'discount_price'] = full_data_drop.loc[USD_index, 'discount_price'] * 0.84
print(full_data_drop.loc[USD_index, 'discount_price'].head(3))
```

```python
full_data_drop.loc[CHF_index, 'shipping_option_price'] = full_data_drop.loc[CHF_index, 'shipping_option_price'] * 0.91
full_data_drop.loc[USD_index, 'shipping_option_price'] = full_data_drop.loc[USD_index, 'shipping_option_price'] * 0.84
```
